## Using Naive Bayes

```python
from sklearn.model_selection import train_test_split
#DROPPING OUT TARGET AND TRESTBPS BEACUSE OF THEIR PRECISE VALUES
```

```python
X=df.drop(['target','trestbps' , 'thalach'],axis=1)
Y=df["target"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB #IMPORTINGNAIVE BAYES ALGO
```

```python
algo=GaussianNB()
```

```python

algo.fit(X_train,Y_train)
```

```python
Y_predict=algo.predict(X_test)
```

```python
Y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, Y_predict)
```

```python
#THE NAIVE BAYES ALGO GIVES US A ACCURACY SCORE OF 0.81
```
