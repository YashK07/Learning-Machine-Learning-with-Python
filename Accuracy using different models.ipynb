```python

#packages from data analysis
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop('Cabin', axis=1,inplace=True)
data.dropna(inplace=True)
```

```python
data.head(10)
```

```python

data.isnull().sum()
```

```python
#hence no null values
```

```python

data['Male']=pd.get_dummies(data['Sex'],drop_first=True)
```

```python
data.drop(['Name','Sex'],inplace=True,axis=1)
```

```python
data.head(10)
```

```python
pd.get_dummies(data['Embarked'],drop_first=True)
Embarked=pd.get_dummies(data['Embarked'],drop_first=True)
Embarked.head(5)
```

```python
data.drop(['PassengerId'],inplace=True,axis=1)
```

```python
data=pd.concat((data,Embarked),axis=1)
```

```python
data.drop(['Ticket','Embarked'],inplace=True,axis=1)
```

```python
data.head(10)
```

using different models 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import neighbors
from sklearn.tree import DecisionTreeClassifier
```

```python

Lr=LogisticRegression()
Gb=GaussianNB()
KNN=neighbors.KNeighborsClassifier(n_neighbors=8)
Dtc=DecisionTreeClassifier()
```

```python
X=data.drop(['Survived'], axis=1)
Y=data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

SPLITTING DATASET

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=1)
```

```python
Lr.fit(X_train,Y_train)
```

```python
Dtc.fit(X_train,Y_train)
```

```python
KNN.fit(X_train,Y_train)
```

```python
Gb.fit(X_train,Y_train)
```

```python
Y_predict1=Lr.predict(X_test)
Y_predict2=Gb.predict(X_test)
Y_predict3=KNN.predict(X_test)
Y_predict4=Dtc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score1=accuracy_score(Y_test,Y_predict1)
accuracy_score2=accuracy_score(Y_test,Y_predict2)
accuracy_score3=accuracy_score(Y_test,Y_predict3)
accuracy_score4=accuracy_score(Y_test,Y_predict4)
```
