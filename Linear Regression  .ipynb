```python
#packages from data analysis
import numpy as np
import pandas as pd

#for the simple vector machine
from sklearn import svm

#for charts and visualisation
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)

%matplotlib inline
```

```python
#reading the data
```

```python
df=pd.read_csv('index.csv')
print(df.shape)
df.head(60)
```

```python
X = df['Weight'].values
Y = df['Index'].values
mean_x = np.mean(X)
mean_y = np.mean(Y)
m = len(X)
numer = 0
denom = 0
for i in range(m):
        numer += (X[i] - mean_x) * (Y[i] - mean_y)
        denom += (X[i] - mean_x) ** 2
b1 = numer/denom
b0 = mean_y - (b1 * mean_x)
print(b1,b0)
```

```python
#plotting all values 
max_x=np.max(X)+100
min_x=np.min(X)-100
#reg line
x=np.linspace(min_x,max_x,1000)    #to get points on X axis for the line
y= b0 + b1*x                      #reg line

plt.plot(x,y,color='#58b970',label='Regression Line')

plt.scatter(X,Y,color='#ef5423',label='Scatter Plot')


plt.xlabel('Weight')
plt.ylabel('Index')
plt.title('Linear Regression on wine dataset')
plt.legend()
plt.show()
```

```python

tss=0
tsr=0
for i in range(m):
    ypred=b1*X[i]+ b0
    tss+=(Y[i]-mean_y)**2
    tsr+=(Y[i]-ypred)**2
r2= 1-tsr/tss
print(r2)
```

```python

from sklearn.linear_model import LinearRegression
```

```python

X=X.reshape(m,1)
lr=LinearRegression()
fit=lr.fit(X,Y)
ypred2=lr.predict(X) #predicting y
r2_score=lr.score(X,Y) #calculating r2 score
```
